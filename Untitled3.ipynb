```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn import metrics
from sklearn.metrics import r2_score
from sklearn.model_selection import KFold
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
%matplotlib inline

from sklearn.linear_model import LinearRegression
from sklearn import svm
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.tree import DecisionTreeRegressor



dataset = pd.read_csv('E:/4.1/AI/Project/CPUfinal.csv')



#let's see the distribution of price

sns.distplot(dataset['Price'],bins = 20)

#let's see the correlation of the attributes

ax = plt.subplots(figsize = (14,6))
sns.heatmap(dataset.corr(),cmap = 'magma',linecolor = 'white',lw = 1)

#we intend to use features that have good correlation with the target variable

col=["Cspeed","Cache","Cores","HT","Unlocked","MCS"]
X=dataset[col]
y=dataset["Price"]

#let's scale the features

MMS = MinMaxScaler()
X=MMS.fit_transform(X)

r2_lr=[]
mae_lr=[]
mse_lr=[]
rmse_lr=[]

r2_svr=[]
mae_svr=[]
mse_svr=[]
rmse_svr=[]

r2_dt=[]
mae_dt=[]
mse_dt=[]
rmse_dt=[]

r2_rf=[]
mae_rf=[]
mse_rf=[]
rmse_rf=[]

r2_gb=[]
mae_gb=[]
mse_gb=[]
rmse_gb=[]

LinReg = LinearRegression()
SVMR = svm.SVR()
RFR = RandomForestRegressor(max_depth=10, random_state=0)
DTR = DecisionTreeRegressor(random_state=0)
GBR = GradientBoostingRegressor()

kf = KFold(n_splits=5,random_state = 0,shuffle = True)

for train_index, test_index in kf.split(X):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
    
    LinReg.fit(X_train,y_train)
    pred_y_lr = LinReg.predict(X_test)
    
    r2_lr.append(r2_score(y_test, pred_y_lr))
    mae_lr.append(metrics.mean_absolute_error(y_test, pred_y_lr))
    mse_lr.append(metrics.mean_squared_error(y_test, pred_y_lr))
    rmse_lr.append(np.sqrt(metrics.mean_squared_error(y_test, pred_y_lr)))
    
    SVMR.fit(X_train,y_train,)
    pred_y_svr = SVMR.predict(X_test)
    
    r2_svr.append(r2_score(y_test, pred_y_svr))
    mae_svr.append(metrics.mean_absolute_error(y_test, pred_y_svr))
    mse_svr.append(metrics.mean_squared_error(y_test, pred_y_svr))
    rmse_svr.append(np.sqrt(metrics.mean_squared_error(y_test, pred_y_svr)))
    
    DTR.fit(X_train,y_train)
    pred_y_dt = DTR.predict(X_test)
    
    r2_dt.append(r2_score(y_test, pred_y_dt))
    mae_dt.append(metrics.mean_absolute_error(y_test, pred_y_dt))
    mse_dt.append(metrics.mean_squared_error(y_test, pred_y_dt))
    rmse_dt.append(np.sqrt(metrics.mean_squared_error(y_test, pred_y_dt)))
    
    
    RFR.fit(X_train,y_train,)
    pred_y_rf = RFR.predict(X_test)
    
    r2_rf.append(r2_score(y_test, pred_y_rf))
    mae_rf.append(metrics.mean_absolute_error(y_test, pred_y_rf))
    mse_rf.append(metrics.mean_squared_error(y_test, pred_y_rf))
    rmse_rf.append(np.sqrt(metrics.mean_squared_error(y_test, pred_y_rf)))
    
    GBR.fit(X_train,y_train,)
    pred_y_gb = GBR.predict(X_test)
    
    r2_gb.append(r2_score(y_test, pred_y_gb))
    mae_gb.append(metrics.mean_absolute_error(y_test, pred_y_gb))
    mse_gb.append(metrics.mean_squared_error(y_test, pred_y_gb))
    rmse_gb.append(np.sqrt(metrics.mean_squared_error(y_test, pred_y_gb)))
    
    
    
print ("Linear Regression:")
print ("R2 : ",np.mean(r2_lr))
print ("Mean Absolute Error : ",np.mean(mae_lr))
print ("Mean Squared Error : ",np.mean(mse_lr))
print ("Root MSE  : ",np.mean(rmse_lr))
print ("       ")

print ("Support Vector Regression :")
print ("R2 : ",np.mean(r2_svr))
print ("Mean Absolute Error : ",np.mean(mae_svr))
print ("Mean Squared Error : ",np.mean(mse_svr))
print ("Root MSE  : ",np.mean(rmse_svr))
print ("       ")

print ("Decision Tree regression :")
print ("R2 : ",np.mean(r2_dt))
print ("Mean Absolute Error : ",np.mean(mae_dt))
print ("Mean Squared Error : ",np.mean(mse_dt))
print ("Root MSE  : ",np.mean(rmse_dt))
print ("       ")

print ("Random Forest Regression :")
print ("R2 : ",np.mean(r2_rf))
print ("Mean Absolute Error : ",np.mean(mae_rf))
print ("Mean Squared Error : ",np.mean(mse_rf))
print ("Root MSE  : ",np.mean(rmse_rf))
print ("       ")

print ("Gradient Boosting Regression :")
print ("R2 : ",np.mean(r2_gb))
print ("Mean Absolute Error : ",np.mean(mae_gb))
print ("Mean Squared Error : ",np.mean(mse_gb))
print ("Root MSE  : ",np.mean(rmse_gb))
print ("       ")
```
